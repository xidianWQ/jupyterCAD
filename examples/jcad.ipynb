```python
import os
# 强制禁用 TBB 多线程，改为单线程运行（防止部分环境崩溃）
os.environ["OCCT_NO_TBB"] = "1"

from OCC.Core.BRepPrimAPI import BRepPrimAPI_MakeBox
from OCC.Core.XCAFApp import XCAFApp_Application
from OCC.Core.TDocStd import TDocStd_Document
from OCC.Core.TCollection import TCollection_ExtendedString, TCollection_AsciiString
from OCC.Core.XCAFDoc import XCAFDoc_DocumentTool
from OCC.Core.RWGltf import RWGltf_CafWriter
from OCC.Core.TDataStd import TDataStd_Name
from OCC.Core.TColStd import TColStd_IndexedDataMapOfStringString
# 新增：导入 Message 模块
from OCC.Core.Message import Message_ProgressRange

print("1. 开始创建几何体...")
shape = BRepPrimAPI_MakeBox(10, 20, 30).Shape()

print("2. 初始化 XCAF 文档...")
h_app = XCAFApp_Application.GetApplication()
doc = TDocStd_Document(TCollection_ExtendedString("BinXCAF"))
h_app.NewDocument(TCollection_ExtendedString("BinXCAF"), doc)

print("3. 添加形状到文档...")
shape_tool = XCAFDoc_DocumentTool.ShapeTool(doc.Main())
label = shape_tool.AddShape(shape)
TDataStd_Name.Set(label, TCollection_ExtendedString("MyBox"))

print("4. 开始导出 GLB...")
# 构造 Writer
writer = RWGltf_CafWriter(TCollection_AsciiString("test_pure.glb"), True)

# 准备参数：必须显式传入文件信息和进度条对象
file_info = TColStd_IndexedDataMapOfStringString()
progress = Message_ProgressRange()

# 执行导出 (3个参数)
writer.Perform(doc, file_info, progress)

print("5. 成功！")
```
